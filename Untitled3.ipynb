```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from transformers import BertTokenizer, BertModel

# Load BERT tokenizer and model
model_name = "bert-base-uncased"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name, output_attentions=True)
```

```python
def get_attention_scores(text):
    inputs = tokenizer(text, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)

    attentions = torch.stack(outputs.attentions).squeeze(1).numpy()  # Extract attention scores
    return attentions, inputs
```

```python
def plot_attention_heatmap(text):
    attentions, inputs = get_attention_scores(text)
    tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"].squeeze().tolist())

    # Average attention across all heads and layers
    avg_attention = np.mean(attentions, axis=(0, 1))

    # Plot heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(avg_attention, xticklabels=tokens, yticklabels=tokens, cmap="Blues", annot=True)
    plt.title("Attention Heatmap")
    plt.show()
```

```python
medical_note1 = "Patient reports persistent cough, high fever, and difficulty breathing for the past three days."
plot_attention_heatmap(medical_note1)
```

```python
medical_note2 = "Mild headache and occasional dizziness, but no fever or cough."
plot_attention_heatmap(medical_note2)
```

```python
rare_symptom_note = "Patient experiences sudden vision loss and severe chest tightness."
plot_attention_heatmap(rare_symptom_note)
```
